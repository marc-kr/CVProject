```python
from ultralytics import RTDETR
```

```python
model = RTDETR("rtdetr-l.pt")
for name, module in model.named_children():
    print(name, module)
```

```python
results = model.train(data="/kaggle/input/dataset-football-cv/data.yaml", 
                      epochs=5, 
                      imgsz=1600,
                      save_period=1,
                      plots=True,  
                      exist_ok=True, 
                      augment=True,
                      hsv_h = 0.01,
                      hsv_s=0.0,
                      hsv_v=0.2,
                      shear=2,
                      scale=0.2,
                      mosaic=0.0,
                      seed=42,
                      cls=1.5,
                      device=[0,1]
                      batch=8
                     )
```
